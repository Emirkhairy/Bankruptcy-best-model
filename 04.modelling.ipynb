```python
# Import necessary package

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix, recall_score, precision_score, roc_auc_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from plot_metric.functions import BinaryClassification
```

```python
# assignts the model into a list
models ={'LR': LogisticRegression(),
        'DT': DecisionTreeClassifier(),
        'RF':RandomForestClassifier()}
```

```python
models
```

```python
model_name = list(models.keys())
```

```python
model_name[1]
```

```python
def get_performencs(x_train, x_test, y_train, y_test, model_name):
    """
    Fit and Predict a data
    and find accuracy,precision,recall,roc and F1 score
    
    
    Paramaters
    ---
    x_train,x_test,y_train,y_test: DataFrame
    model_name: str
    
    """
    model = models[model_name]
    model.fit(x_train,y_train)
    
    prediction = model.predict(x_test)
    
    accuracy = accuracy_score(y_test,prediction)*100
    precision = precision_score(y_test,prediction)*100
    recall = recall_score(y_test,prediction)*100
    roc = roc_auc_score(y_test,prediction)*100
    f1 = f1_score(y_test,prediction)*100
    
    return accuracy,precision,recall,roc,f1
```

```python
# Create a blank list to restore the result later
accuracy_list,precision_list,recall_list,roc_list,f1_list,model_list=[],[],[],[],[],[]
for i in model_name:
    accuracy, precision, recall, roc, f1 = get_performencs(x_smote, x_fs, y_smote, y_test, i)
    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    roc_list.append(roc)
    f1_list.append(f1)
    model_list.append(i)

# Create a dictionary to store the lists
data = {
    'Model': model_list,
    'Accuracy': accuracy_list,
    'Precision': precision_list,
    'Recall': recall_list,
    'ROC_AUC': roc_list,
    'F1_Score': f1_list
}

# Create the dataframe from the dictionary
performance_df = pd.DataFrame(data)
```

```python
performance_df
```

```python
# Make a loop for show the ROC score from each model
for m in model_name:
    model = models[m]
    model.fit(x_smote, y_smote)
    y_pred = model.predict(x_fs)
   
    #Binary Classification
    bc = BinaryClassification(y_test, y_pred, labels=["Class 0", "Class 1"])
    plt.figure(figsize=(5, 5))
    bc.plot_roc_curve()
    plt.title(f'ROC Curve for {m}')
    plt.show()
    
```
